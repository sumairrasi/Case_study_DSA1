# 4. Find the top 5 cars based on their sales price.

```python
pop_data1 = pd.DataFrame(data,columns=['Make','Sale Price'])
```

```python
pop_data3 = pop_data1.groupby('Make')['Sale Price'].apply(lambda ser: ser.drop_duplicates().nlargest(1)).droplevel(level=1).sort_index().reset_index()
```

```python
pop_data2 = pop_data3.nlargest(n=5 ,columns=['Sale Price'])
```

```python
pop_data2
```

# 5. Find the least 5 cars based on their Resell price

```python
dop_data1 = pd.DataFrame(data,columns=['Make','Resell Price'])
```

```python
dop_data2 = dop_data1.groupby('Make')['Resell Price'].apply(lambda ser: ser.drop_duplicates().nsmallest(1)).droplevel(level=1).sort_index().reset_index()
```

```python
dop_data3 = dop_data2.nsmallest(n=5 ,columns=['Resell Price'])
```
